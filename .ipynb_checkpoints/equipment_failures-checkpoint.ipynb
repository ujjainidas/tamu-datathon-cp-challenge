```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

equip_data = pd.read_csv("equip_failures_training_set.csv", na_values = "na")
equip_data.head()
```

```python
descriptive_stats = equip_data.describe()
descriptive_stats
```

```python
def cleanData(data_set):
    cleaned_data = data_set
    cleaned_data.drop("id", axis = 1)
    for c in cleaned_data.columns:
        cleaned_data[c] = cleaned_data[c].fillna(0)
    return cleaned_data
```

```python
data_copy = equip_data.copy()
features = data_copy.drop(["target"], 1)
predict = data_copy["target"]

features = cleanData(features)

sc = StandardScaler()
features = pd.DataFrame(sc.fit_transform(features.values), index = features.index, columns = features.columns)

train_X, test_X, train_Y, test_Y = train_test_split(features, predict, test_size = 0.08, random_state = 150)
```

```python
linReg = LogisticRegression(solver='lbfgs', random_state=150)
model = linReg.fit(train_X, train_Y)

print(model.coef_)

prediction = linReg.predict(test_X)
print(prediction)

print(linReg.score(test_X, test_Y))
```
